## Demonstrate the PMTarray functionalities

```python
import pmtarray
```

```python
array = pmtarray.PMTarray(array_diameter=2600,
                          border_margin=-10,
                          intra_pmt_distance= 10.,
                          pmt_model='3in')

array.print_properties()
array.plot_pmt_array()
```

```python
array = pmtarray.PMTarray(array_diameter=500,
                          border_margin=10,
                          intra_pmt_distance= 50.,
                          pmt_model='2in')

array.print_properties()
array.plot_pmt_array()
```

### Custom units

#### Circular

```python
custom_params = {'type': 'circular',
                 'name': 'The BEST PMT',
                 'diameter_packaging': 50,
                 'active_diameter': 20,
                 'diameter_tolerance':0,
                 'qe': 0.34,
                 'active_area_correction': 1.
                 }

array = pmtarray.PMTarray(array_diameter=200,
                          border_margin=10,
                          intra_pmt_distance= 5.,
                          pmt_model='custom',
                          custom_unit_params=custom_params)

array.print_properties()
array.plot_pmt_array()
```

#### Rectangular

```python
custom_params = {'name' : 'The BEST PMT square',
                 'type' : 'square',
                 'width_package' : 50,
                 'height_package' : 50,
                 'width_tolerance' : 0,
                 'height_tolerance' : 0,
                 'width_active' : 20,
                 'height_active' : 20,
                 'active_area_correction' : 1,
                 'D_corner_x_active' : 2.5,
                 'D_corner_y_active' : 2.5,
                 'qe' : 0.24,
                 }

array = pmtarray.PMTarray(array_diameter=300,
                          border_margin=10,
                          intra_pmt_distance= 5.,
                          pmt_model='custom',
                          custom_unit_params=custom_params)

array.print_properties()
array.plot_pmt_array()
```
